```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['workclass'] = encoder.fit_transform(data['workclass'])
data['education'] = encoder.fit_transform(data['education'])
data['marital-status'] = encoder.fit_transform(data['marital-status'])
data['occupation'] = encoder.fit_transform(data['occupation'])
data['relationship'] = encoder.fit_transform(data['relationship'])
data['race'] = encoder.fit_transform(data['race'])
data['gender'] = encoder.fit_transform(data['gender'])
data['native-country'] = encoder.fit_transform(data['native-country'])
data
```

```python
x = data.drop(columns = ['income']) # input data
y = data['income'] # output data
x
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, OneHotEncoder

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) #0.2 = 20% test data and 80% training data

models = {
    "LogisticRegression": LogisticRegression(),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier()
}

results = {}

for name, model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
plt.bar(results.keys(), results.values(), color = 'green')
plt.xlabel('Model')
plt.ylabel('Accuracy score')
plt.xticks(rotation=45)
plt.title('Model Accuracy Comparison')
plt.grid(True)
plt.show()
```

```python
import joblib
# finding the best model

best_model_name = max(results, key=results.get)
best_model = models[best_model_name]
print(f"\n Best model: {best_model_name} with accuracy {results[best_model_name]:.4f}")

# Save the best model
joblib.dump(best_model, 'best_model.pkl')
print("Saved best model as best_model.pkl ")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder, StandardScaler # Keep LabelEncoder for transforming input

# Load the trained model
model = joblib.load("best_model.pkl")

# Load the original data to fit encoders and scaler
# Make sure the path to your data file is correct
try:
    original_data = pd.read_csv('/content/adult.csv')
except FileNotFoundError:
    st.error("Error: adult.csv not found. Please make sure the file is in the correct path.")
    st.stop()

# Handle '?' replacements in the original data as done during training
original_data['workclass'] = original_data['workclass'].replace({'?':'notlisted'})
original_data['occupation'] = original_data['occupation'].replace({'?':'others'})
original_data['native-country'] = original_data['native-country'].replace({'?':'others'})

st.set_page_config(page_title="Employee Salary Prediction", page_icon="💼", layout="wide")

st.title("💼 Employee Salary Prediction App")
st.markdown("Predict whether an employee earns >50K or ≤50K based on input features.")

# Sidebar inputs (these must match your training feature columns)
st.sidebar.header("Input Employee Details")

# Input fields based on user's request and available training data
# Use the range and unique values from the original data for the input widgets
age = st.sidebar.slider("Age", int(original_data['age'].min()), int(original_data['age'].max()), 30)
workclass = st.sidebar.selectbox("Workclass", original_data['workclass'].unique())
# Updated education options to exclude removed values
education_options = original_data['education'].unique().tolist()
education_options_to_remove = ['Preschool', '1st-4th', '5th-6th']
education_options = [item for item in education_options if item not in education_options_to_remove]
education = st.sidebar.selectbox("Education", education_options)

occupation = st.sidebar.selectbox("Occupation", original_data['occupation'].unique())
marital_status = st.sidebar.selectbox("Marital Status", original_data['marital-status'].unique())
gender = st.sidebar.selectbox("Gender", original_data['gender'].unique())
hours_per_week = st.sidebar.slider("Hours per week", int(original_data['hours-per-week'].min()), int(original_data['hours-per-week'].max()), 40)


# Build input DataFrame (⚠️ must match preprocessing of your training data)
input_df = pd.DataFrame({
    'age': [age],
    'workclass': [workclass],
    'fnlwgt': [original_data['fnlwgt'].mean()], # Using mean of original data as a placeholder
    'education': [education],
    'educational-num': [original_data['educational-num'].mean()], # Using mean of original data as a placeholder
    'marital-status': [marital_status],
    'occupation': [occupation],
    'relationship': [original_data['relationship'].mode()[0]], # Using mode of original data as a placeholder
    'race': [original_data['race'].mode()[0]], # Using mode of original data as a placeholder
    'gender': [gender],
    'capital-gain': [original_data['capital-gain'].mean()], # Using mean of original data as a placeholder
    'capital-loss': [original_data['capital-loss'].mean()], # Using mean of original data as a placeholder
    'hours-per-week': [hours_per_week],
    'native-country': [original_data['native-country'].mode()[0]] # Using a placeholder mode from original data
})

# Apply the same label encoding as during training
categorical_cols = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country']
for col in categorical_cols:
    encoder = LabelEncoder()
    # Fit the encoder on the unique values from the original data
    # Ensure the encoder is fitted on the data *after* removing the values you excluded
    filtered_original_data = original_data[~original_data['education'].isin(education_options_to_remove)]
    encoder.fit(filtered_original_data[col].unique())
    input_df[col] = encoder.transform(input_df[col])


st.write("### 🔎 Input Data after Encoding")
st.write(input_df)


# Predict button
if st.button("Predict Salary Class"):
    # Scale the input data before predicting
    scaler = StandardScaler()
    numerical_cols = ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']
    # Fit the scaler on the numerical columns of the original data
    scaler.fit(original_data[numerical_cols])

    input_df_scaled = input_df.copy()
    input_df_scaled[numerical_cols] = scaler.transform(input_df[numerical_cols])


    prediction = model.predict(input_df_scaled)
    st.success(f"✅ Prediction: {prediction[0]}")

# Batch prediction
st.markdown("---")
st.markdown("#### 📂 Batch Prediction")
uploaded_file = st.file_uploader("Upload a CSV file for batch prediction", type="csv")

if uploaded_file is not None:
    batch_data = pd.read_csv(uploaded_file)

    # Handle '?' replacements in the batch data
    batch_data['workclass'] = batch_data['workclass'].replace({'?':'notlisted'})
    batch_data['occupation'] = batch_data['occupation'].replace({'?':'others'})
    batch_data['native-country'] = batch_data['native-country'].replace({'?':'others'})


    # Apply label encoding
    for col in categorical_cols:
        encoder = LabelEncoder()
        # Fit the encoder on the unique values from the original data
        # Ensure the encoder is fitted on the data *after* removing the values you excluded
        filtered_original_data = original_data[~original_data['education'].isin(education_options_to_remove)]
        encoder.fit(filtered_original_data[col].unique())
        batch_data[col] = encoder.transform(batch_data[col])

    # Scale numerical features
    scaler = StandardScaler()
    numerical_cols = ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']
    # Fit the scaler on the numerical columns of the original data
    scaler.fit(original_data[numerical_cols])
    batch_data_scaled = batch_data.copy()
    batch_data_scaled[numerical_cols] = scaler.transform(batch_data[numerical_cols])

    st.write("Uploaded data preview:", batch_data.head())
    batch_preds = model.predict(batch_data_scaled)
    batch_data['PredictedClass'] = batch_preds
    st.write("✅ Predictions:")
    st.write(batch_data.head())
    csv = batch_data.to_csv(index=False).encode('utf-8')
    st.download_button("Download Predictions CSV", csv, file_name='predicted_classes.csv', mime='text/csv')
```
